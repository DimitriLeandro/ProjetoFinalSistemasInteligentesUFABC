Pacotes que serão usados para importação do *dataset*, exibição de gráficos, separação de dados entre treinamento e teste, método de redução de dimensionalidade e método de clusterização

```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
```

## Pré-Processamento dos dados

Importação do dataset *Wine*.

```python
dataset = datasets.load_wine()
```

Atribuição as variáveis X e y os dados e o elemento *target*, respectivamente.

```python
X = dataset.data
y = dataset.target
```

Imprimindo o formato de ambas variáveis geradas.

```python
X.shape
```

```python
y.shape
```

Podemos perceber que estamos trabalhando com um *dataset* que possui 13 classes, e isso talvez seja difícil de visualizar durante a clusterização dos dados, por esse motivo iremos reduzir as dimensionalidades do *dataset* através do método de redução de dimensionalidade PCA. Guardaremos esses novos dados na variável X_pca.

```python
X_pca = PCA(n_components = 2).fit_transform(X)
```

Podemos visualizar que o número de classes dos dados foram alterados.

```python
X_pca.shape
```

Em seguida vamos separar os dados entre dados de treinamento e dados de teste utilizando a função **train_test_split**, utilizaremos 25% dos dados para teste.

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size = 0.25) 
```

Abaixo temos o formato dos vetores de teste e treinamento gerados.

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

## K-Means SK-Learn

Inicialmente vamos definir o clusterizador com nome *kmeans* que recebe a função "*KMeans*" e tem como argumento o número de clusters a serem gerados pelo algoritmo K-Means, utilizaremos *n_clusters*=3.

```python
kmeans = KMeans(n_clusters=3)
```

Após isso treinaremos nosso clusterizador com os dados de treinamento.

```python
kmeans.fit(X_train)
```

E prediremos conforme os dados de teste.

```python
y_pred = kmeans.predict(X_test)
```

Podemos exibir os centros de clusters que foram gerados através da função "*cluster_centers_*".

```python
kmeans.cluster_centers_
```

E plotar o gráfico mostrando a distribuição dos dados e as respectivas classificações em clusters realizadas pelo método de clusterização.

```python
plt.figure()
plt.scatter(x=X_test[:, 0], y=X_test[:, 1], c=y_pred, cmap='viridis', s=200, alpha=0.5)
plt.xlim(-500, 900)
plt.ylim(-20, 25)
plt.xlabel('Eixo PCA 1')
plt.ylabel('Eixo PCA 2')
plt.title('K-Means Clusters')
plt.grid(True)

plt.scatter(x=kmeans.cluster_centers_[:,0], y=kmeans.cluster_centers_[:,1], c='r', s=50, marker='X', label='Centroides')
plt.legend()
```

Como métrica do clusterizador já implementado pelo SK-Learn temos apenas a inércia.

```python
kmeans.inertia_
```
